```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 3: Imports & GPU
# ══════════════════════════════════════════════════════════════════════════════
import json, random, time, gc
from collections import Counter
import numpy as np
import pandas as pd
from PIL import Image
import torch
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from tqdm.auto import tqdm

MODEL_ID = "HuggingFaceTB/SmolVLM-500M-Instruct"
CHOICE_LETTERS = "ABCDEFGHIJ"
MAX_CONTEXT_CHARS = 400

TRAIN_BATCH = 4
GRAD_ACCUM = 2
INFER_BATCH = 16
NUM_WORKERS = 4

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
if torch.cuda.is_available():
    gpu_name = torch.cuda.get_device_name(0)
    vram = torch.cuda.get_device_properties(0).total_memory / 1e9
    USE_BF16 = torch.cuda.is_bf16_supported()
    COMPUTE_DTYPE = torch.bfloat16 if USE_BF16 else torch.float16
    print(f"GPU: {gpu_name} | VRAM: {vram:.1f}GB | dtype: {COMPUTE_DTYPE}")
else:
    COMPUTE_DTYPE = torch.float32
    print("WARNING: No GPU")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 4: Load Data + Captions
# ══════════════════════════════════════════════════════════════════════════════
train_df = pd.read_csv(DATA_DIR / "train.csv")
val_df   = pd.read_csv(DATA_DIR / "val.csv")
test_df  = pd.read_csv(DATA_DIR / "test.csv")

for df in [train_df, val_df, test_df]:
    df["choices"] = df["choices"].apply(json.loads)

combined_df = pd.concat([train_df, val_df], ignore_index=True)
print(f"Combined: {len(combined_df)} | Test: {len(test_df)}")

# Captions
caption_file = CAPTION_DIR / "all_captions.json"
assert caption_file.exists(), f"Captions not found at {caption_file}"
with open(caption_file) as f:
    all_captions = json.load(f)
print(f"Loaded {len(all_captions)} captions.")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 5: Prompt Builders
# ══════════════════════════════════════════════════════════════════════════════

def _trunc(text, max_chars=MAX_CONTEXT_CHARS):
    if pd.isna(text): return ""
    text = str(text).strip()
    if len(text) <= max_chars: return text
    return text[:max_chars].rsplit(" ", 1)[0] + "..."


def prompt_caption_meta(row, captions, include_answer=False):
    choices = row["choices"]
    choices_str = "\n".join(f"  {CHOICE_LETTERS[i]}. {c}" for i, c in enumerate(choices))
    lecture = _trunc(row.get("lecture", ""))
    hint = _trunc(row.get("hint", ""))
    caption = captions.get(str(row["id"]), "")
    subject = str(row.get("subject", "")).strip() if pd.notna(row.get("subject")) else ""
    grade = str(row.get("grade", "")).strip() if pd.notna(row.get("grade")) else ""

    context_parts = []
    if caption: context_parts.append(f"Image description: {caption[:300]}")
    if lecture: context_parts.append(lecture)
    if hint: context_parts.append(hint)
    context_str = "\n".join(context_parts)

    prompt = "<image>\n"
    if subject or grade:
        meta = []
        if subject: meta.append(f"Subject: {subject}")
        if grade: meta.append(f"Grade: {grade}")
        prompt += " | ".join(meta) + "\n\n"
    if context_str:
        prompt += f"Context:\n{context_str}\n\n"
    prompt += f"Question: {row['question']}\n"
    prompt += f"Choices:\n{choices_str}\n"
    prompt += "Answer:"
    if include_answer:
        prompt += f" {CHOICE_LETTERS[int(row['answer'])]}"
    return prompt


def prompt_cot(row, captions, include_answer=False):
    """Chain-of-thought style prompt — structurally different."""
    choices = row["choices"]
    choices_str = "\n".join(f"  {CHOICE_LETTERS[i]}. {c}" for i, c in enumerate(choices))
    caption = captions.get(str(row["id"]), "")
    lecture = _trunc(row.get("lecture", ""))
    hint = _trunc(row.get("hint", ""))
    subject = str(row.get("subject", "")).strip() if pd.notna(row.get("subject")) else ""

    prompt = "<image>\n"
    if subject:
        prompt += f"This is a {subject} question.\n\n"
    if caption:
        prompt += f"Image shows: {caption[:250]}\n"
    if lecture:
        prompt += f"Background: {lecture}\n"
    if hint:
        prompt += f"Hint: {hint}\n"
    prompt += f"\nQ: {row['question']}\n{choices_str}\nBest answer letter:"
    if include_answer:
        prompt += f" {CHOICE_LETTERS[int(row['answer'])]}"
    return prompt


print("Prompt builders ready.")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 6: Dataset + FIXED Collator (tile padding)
# ══════════════════════════════════════════════════════════════════════════════

class TrainDataset(Dataset):
    def __init__(self, df, processor, data_dir, captions, prompt_fn):
        self.df = df.reset_index(drop=True)
        self.processor = processor
        self.data_dir = data_dir
        self.captions = captions
        self.prompt_fn = prompt_fn

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        image = Image.open(self.data_dir / row["image_path"]).convert("RGB")

        full_text = self.prompt_fn(row, self.captions, include_answer=True)
        prompt_text = self.prompt_fn(row, self.captions, include_answer=False)

        full_enc = self.processor(
            text=[full_text], images=[image],
            return_tensors="pt", truncation=False
        )
        prompt_enc = self.processor(
            text=[prompt_text], images=[image],
            return_tensors="pt", truncation=False
        )

        return {
            "input_ids": full_enc["input_ids"].squeeze(0),
            "attention_mask": full_enc["attention_mask"].squeeze(0),
            "pixel_values": full_enc["pixel_values"].squeeze(0),
            "prompt_len": prompt_enc["input_ids"].shape[1],
        }


def collate_fn(batch, pad_id):
    """Left-pad token sequences; pad tile dimension of pixel_values."""
    max_len = max(x["input_ids"].shape[0] for x in batch)
    max_tiles = max(x["pixel_values"].shape[0] for x in batch)
    ids, masks, labels, pixels, pixel_masks = [], [], [], [], []

    for x in batch:
        seq_len = x["input_ids"].shape[0]
        pad_len = max_len - seq_len

        # Token padding (left-pad)
        padded_ids = torch.cat([
            torch.full((pad_len,), pad_id, dtype=x["input_ids"].dtype),
            x["input_ids"]
        ])
        padded_mask = torch.cat([
            torch.zeros(pad_len, dtype=x["attention_mask"].dtype),
            x["attention_mask"]
        ])

        lab = torch.full_like(padded_ids, -100)
        answer_start = pad_len + x["prompt_len"]
        lab[answer_start:] = padded_ids[answer_start:]

        # Tile padding for pixel_values
        num_tiles, C, H, W = x["pixel_values"].shape
        tile_pad = max_tiles - num_tiles
        if tile_pad > 0:
            pv = torch.cat([
                x["pixel_values"],
                torch.zeros(tile_pad, C, H, W, dtype=x["pixel_values"].dtype)
            ], dim=0)
            real_mask = torch.ones(num_tiles, H, W, dtype=torch.bool)
            pad_mask = torch.zeros(tile_pad, H, W, dtype=torch.bool)
            pm = torch.cat([real_mask, pad_mask], dim=0)
        else:
            pv = x["pixel_values"]
            pm = torch.ones(num_tiles, H, W, dtype=torch.bool)

        ids.append(padded_ids)
        masks.append(padded_mask)
        labels.append(lab)
        pixels.append(pv)
        pixel_masks.append(pm)

    return {
        "input_ids": torch.stack(ids),
        "attention_mask": torch.stack(masks),
        "labels": torch.stack(labels),
        "pixel_values": torch.stack(pixels),
        "pixel_attention_mask": torch.stack(pixel_masks),
    }

print("Dataset + Fixed Collator ready.")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 7: Model Loader
# ══════════════════════════════════════════════════════════════════════════════
from transformers import AutoProcessor, AutoModelForVision2Seq, BitsAndBytesConfig
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training, TaskType

def load_model(seed, lora_r=8, lora_alpha=16, use_dora=False,
               targets=None):
    if targets is None:
        targets = ["q_proj", "k_proj", "v_proj", "o_proj",
                   "gate_proj", "up_proj", "down_proj"]

    random.seed(seed); np.random.seed(seed)
    torch.manual_seed(seed); torch.cuda.manual_seed_all(seed)

    processor = AutoProcessor.from_pretrained(MODEL_ID)
    if processor.tokenizer.pad_token is None:
        processor.tokenizer.pad_token = processor.tokenizer.eos_token

    bnb_config = BitsAndBytesConfig(
        load_in_4bit=True, bnb_4bit_quant_type="nf4",
        bnb_4bit_compute_dtype=COMPUTE_DTYPE, bnb_4bit_use_double_quant=True,
    )
    model = AutoModelForVision2Seq.from_pretrained(
        MODEL_ID, quantization_config=bnb_config,
        device_map="auto", low_cpu_mem_usage=True,
    )
    model = prepare_model_for_kbit_training(model)

    lora_config = LoraConfig(
        r=lora_r, lora_alpha=lora_alpha, lora_dropout=0.05,
        target_modules=targets, bias="none",
        task_type=TaskType.CAUSAL_LM, use_dora=use_dora,
    )
    model = get_peft_model(model, lora_config)

    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    d = "DoRA" if use_dora else "LoRA"
    ok = "OK" if trainable <= 5_000_000 else "OVER 5M!"
    print(f"{d} r={lora_r} a={lora_alpha} | {trainable:,} ({trainable/1e6:.2f}M) | {ok}")
    return model, processor

print("Model loader ready.")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 8: Training Function (supports flat + cosine LR, saves EVERY epoch)
# ══════════════════════════════════════════════════════════════════════════════
from torch.optim.lr_scheduler import CosineAnnealingLR

def train_model(model, processor, train_df, captions, prompt_fn,
                exp_name, seed=42, epochs=3, lr=2e-4, use_cosine=False):

    print(f"\n{'='*60}")
    print(f"TRAINING: {exp_name}")
    sched_str = "cosine" if use_cosine else "flat"
    print(f"LR={lr} ({sched_str}) | epochs={epochs} | batch={TRAIN_BATCH}x{GRAD_ACCUM}")
    print(f"{'='*60}")

    random.seed(seed); np.random.seed(seed)
    torch.manual_seed(seed); torch.cuda.manual_seed_all(seed)

    pad_id = processor.tokenizer.pad_token_id
    dataset = TrainDataset(train_df, processor, DATA_DIR, captions, prompt_fn)
    loader = DataLoader(
        dataset, batch_size=TRAIN_BATCH, shuffle=True,
        collate_fn=lambda b: collate_fn(b, pad_id),
        num_workers=NUM_WORKERS, pin_memory=True,
    )

    optimizer = torch.optim.AdamW(
        [p for p in model.parameters() if p.requires_grad],
        lr=lr, weight_decay=0.01 if use_cosine else 0.0,
    )

    scheduler = None
    if use_cosine:
        scheduler = CosineAnnealingLR(
            optimizer, T_max=len(loader) * epochs, eta_min=lr * 0.05
        )

    print(f"Samples: {len(dataset)} | Batches/ep: {len(loader)}")

    model.train()
    log = []
    start_time = time.time()
    epoch_models = []  # track saved epoch checkpoints

    for epoch in range(epochs):
        epoch_losses = []
        optimizer.zero_grad()

        pbar = tqdm(enumerate(loader), total=len(loader),
                    desc=f"Ep {epoch+1}/{epochs}", leave=True)

        for batch_idx, batch in pbar:
            batch = {k: v.to(model.device) if torch.is_tensor(v) else v
                     for k, v in batch.items()}

            outputs = model(**batch)
            loss = outputs.loss / GRAD_ACCUM
            loss.backward()
            epoch_losses.append(outputs.loss.item())

            if (batch_idx + 1) % GRAD_ACCUM == 0 or (batch_idx + 1) == len(loader):
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                optimizer.step()
                if scheduler: scheduler.step()
                optimizer.zero_grad()

            avg = np.mean(epoch_losses[-50:])
            cur_lr = scheduler.get_last_lr()[0] if scheduler else lr
            pbar.set_postfix({"loss": f"{outputs.loss.item():.4f}", "avg": f"{avg:.4f}",
                              "lr": f"{cur_lr:.2e}"})

            if (batch_idx + 1) % 200 == 0:
                elapsed = (time.time() - start_time) / 60
                log.append({"ep": epoch+1, "b": batch_idx+1,
                            "loss": outputs.loss.item(), "avg": avg})
                print(f"  [{elapsed:.0f}m] Ep{epoch+1} b{batch_idx+1}/{len(loader)} loss={avg:.4f}")

        epoch_avg = np.mean(epoch_losses)
        print(f"Epoch {epoch+1} — Avg loss: {epoch_avg:.4f}")

        # SAVE EVERY EPOCH (epoch checkpoint harvesting)
        ep_name = f"{exp_name}_ep{epoch+1}"
        ep_path = CHECKPOINT_DIR / ep_name
        model.save_pretrained(str(ep_path))
        epoch_models.append(ep_name)
        print(f"  Saved: {ep_name}")

    elapsed = (time.time() - start_time) / 60
    final_path = CHECKPOINT_DIR / f"{exp_name}_final"
    model.save_pretrained(str(final_path))

    with open(CHECKPOINT_DIR / f"{exp_name}_log.json", "w") as f:
        json.dump(log, f, indent=2)

    print(f"Done: {exp_name} | {elapsed:.1f} min | Epoch checkpoints: {epoch_models}")

    del optimizer, scheduler, loader, dataset
    gc.collect(); torch.cuda.empty_cache()

    return model, epoch_models

print("Training function ready (flat + cosine, epoch harvesting).")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 9: Inference (fixed padding, saves raw scores)
# ══════════════════════════════════════════════════════════════════════════════

def get_candidate_ids(tokenizer):
    cids = {}
    for letter in CHOICE_LETTERS:
        ids = set()
        for form in [letter, " " + letter]:
            enc = tokenizer.encode(form, add_special_tokens=False)
            if enc: ids.add(enc[-1])
        cids[letter] = sorted(ids)
    return cids


def predict_batch(model, processor, df_batch, captions, prompt_fn, cand_ids):
    images = [Image.open(DATA_DIR / row["image_path"]).convert("RGB")
              for _, row in df_batch.iterrows()]
    prompts = [prompt_fn(row, captions, include_answer=False)
               for _, row in df_batch.iterrows()]

    inputs = processor(text=prompts, images=images, return_tensors="pt", padding=True)
    inputs = {k: v.to(model.device) if torch.is_tensor(v) else v
              for k, v in inputs.items()}

    with torch.inference_mode():
        logits = model(**inputs).logits

    seq_lengths = inputs["attention_mask"].sum(dim=1)
    preds, scores_all = [], []

    for i, (_, row) in enumerate(df_batch.iterrows()):
        last_pos = int(seq_lengths[i].item()) - 1
        lp = torch.log_softmax(logits[i, last_pos, :], dim=-1)
        scores = [max(lp[tid].item() for tid in cand_ids[CHOICE_LETTERS[ci]])
                  for ci in range(len(row["choices"]))]
        preds.append(int(np.argmax(scores)))
        scores_all.append(scores)

    return preds, scores_all


def run_inference(model, processor, df, captions, prompt_fn, desc="Infer"):
    model.eval()
    cand_ids = get_candidate_ids(processor.tokenizer)
    all_preds, all_scores = [], []

    pbar = tqdm(range(0, len(df), INFER_BATCH), desc=desc, leave=True)
    for start in pbar:
        end = min(start + INFER_BATCH, len(df))
        preds, scores = predict_batch(model, processor, df.iloc[start:end],
                                       captions, prompt_fn, cand_ids)
        all_preds.extend(preds)
        all_scores.extend(scores)
        pbar.set_postfix({"done": len(all_preds)})
        torch.cuda.empty_cache()

    return all_preds, all_scores


def run_inference_from_checkpoint(ckpt_name, prompt_fn, desc=None):
    """Load a saved checkpoint and run test inference."""
    from peft import PeftModel

    results_path = CHECKPOINT_DIR / f"{ckpt_name}_results.json"
    if results_path.exists():
        print(f"CACHED: {ckpt_name}")
        with open(results_path) as f:
            return json.load(f)

    bnb_config = BitsAndBytesConfig(
        load_in_4bit=True, bnb_4bit_quant_type="nf4",
        bnb_4bit_compute_dtype=COMPUTE_DTYPE, bnb_4bit_use_double_quant=True,
    )
    base = AutoModelForVision2Seq.from_pretrained(
        MODEL_ID, quantization_config=bnb_config,
        device_map="auto", low_cpu_mem_usage=True,
    )
    model = PeftModel.from_pretrained(base, str(CHECKPOINT_DIR / ckpt_name))
    model.eval()

    processor = AutoProcessor.from_pretrained(MODEL_ID)
    if processor.tokenizer.pad_token is None:
        processor.tokenizer.pad_token = processor.tokenizer.eos_token

    preds, scores = run_inference(
        model, processor, test_df, all_captions, prompt_fn,
        desc=desc or ckpt_name[:25]
    )

    result = {"name": ckpt_name, "test_preds": preds, "test_scores": scores}
    with open(results_path, "w") as f:
        json.dump(result, f)

    sub = pd.DataFrame({"id": test_df["id"], "answer": preds})
    sub.to_csv(SUBMISSION_DIR / f"sub_{ckpt_name}.csv", index=False)

    del model, base, processor
    gc.collect(); torch.cuda.empty_cache()

    return result

print("Inference ready (with checkpoint loading).")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 10: Model F — High capacity, cosine scheduler, 4 epochs
# ══════════════════════════════════════════════════════════════════════════════

cfg_F = {
    "name": "F_cap_meta_dora_r16a32_cos_ep4_s42",
    "seed": 42, "lora_r": 16, "lora_alpha": 32,
    "use_dora": True, "prompt_fn": prompt_caption_meta,
    "lr": 2e-4, "epochs": 4, "use_cosine": True,
}

results_path_F = CHECKPOINT_DIR / f"{cfg_F['name']}_final_results.json"

if results_path_F.exists():
    print(f"Model F already complete — skipping training.")
else:
    model_F, processor_F = load_model(
        seed=cfg_F["seed"], lora_r=cfg_F["lora_r"],
        lora_alpha=cfg_F["lora_alpha"], use_dora=cfg_F["use_dora"],
    )
    model_F, F_epochs = train_model(
        model_F, processor_F, combined_df, all_captions,
        prompt_fn=cfg_F["prompt_fn"], exp_name=cfg_F["name"],
        seed=cfg_F["seed"], epochs=cfg_F["epochs"],
        lr=cfg_F["lr"], use_cosine=cfg_F["use_cosine"],
    )

    # Test inference on final model
    print("\nTest inference (Model F final)...")
    preds_F, scores_F = run_inference(
        model_F, processor_F, test_df, all_captions,
        cfg_F["prompt_fn"], desc="Test F_final"
    )
    result_F = {"name": f"{cfg_F['name']}_final", "test_preds": preds_F, "test_scores": scores_F}
    with open(results_path_F, "w") as f:
        json.dump(result_F, f)

    sub = pd.DataFrame({"id": test_df["id"], "answer": preds_F})
    sub.to_csv(SUBMISSION_DIR / f"sub_{cfg_F['name']}_final.csv", index=False)

    del model_F, processor_F
    gc.collect(); torch.cuda.empty_cache()
    print("Model F done.")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 11: Infer on Model F epoch checkpoints
# ══════════════════════════════════════════════════════════════════════════════

F_epoch_results = []
for ep in range(1, cfg_F["epochs"] + 1):
    ckpt = f"{cfg_F['name']}_ep{ep}"
    if (CHECKPOINT_DIR / ckpt).exists():
        print(f"\nInference on {ckpt}...")
        r = run_inference_from_checkpoint(ckpt, cfg_F["prompt_fn"], desc=f"F_ep{ep}")
        F_epoch_results.append(r)
    else:
        print(f"Checkpoint {ckpt} not found — skipping.")

# Also load final
final_rpath = CHECKPOINT_DIR / f"{cfg_F['name']}_final_results.json"
if final_rpath.exists():
    with open(final_rpath) as f:
        F_epoch_results.append(json.load(f))

print(f"\nModel F: {len(F_epoch_results)} checkpoints ready for ensemble.")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 12: Model G — CoT prompt, different reasoning structure
# ══════════════════════════════════════════════════════════════════════════════

cfg_G = {
    "name": "G_cot_dora_r8a16_flat_ep3_s777",
    "seed": 777, "lora_r": 8, "lora_alpha": 16,
    "use_dora": True, "prompt_fn": prompt_cot,
    "lr": 2e-4, "epochs": 3, "use_cosine": False,
}

results_path_G = CHECKPOINT_DIR / f"{cfg_G['name']}_final_results.json"

if results_path_G.exists():
    print(f"Model G already complete — skipping training.")
else:
    model_G, processor_G = load_model(
        seed=cfg_G["seed"], lora_r=cfg_G["lora_r"],
        lora_alpha=cfg_G["lora_alpha"], use_dora=cfg_G["use_dora"],
    )
    model_G, G_epochs = train_model(
        model_G, processor_G, combined_df, all_captions,
        prompt_fn=cfg_G["prompt_fn"], exp_name=cfg_G["name"],
        seed=cfg_G["seed"], epochs=cfg_G["epochs"],
        lr=cfg_G["lr"], use_cosine=cfg_G["use_cosine"],
    )

    print("\nTest inference (Model G final)...")
    preds_G, scores_G = run_inference(
        model_G, processor_G, test_df, all_captions,
        cfg_G["prompt_fn"], desc="Test G_final"
    )
    result_G = {"name": f"{cfg_G['name']}_final", "test_preds": preds_G, "test_scores": scores_G}
    with open(results_path_G, "w") as f:
        json.dump(result_G, f)

    sub = pd.DataFrame({"id": test_df["id"], "answer": preds_G})
    sub.to_csv(SUBMISSION_DIR / f"sub_{cfg_G['name']}_final.csv", index=False)

    del model_G, processor_G
    gc.collect(); torch.cuda.empty_cache()
    print("Model G done.")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 13: Infer on Model G epoch checkpoints
# ══════════════════════════════════════════════════════════════════════════════

G_epoch_results = []
for ep in range(1, cfg_G["epochs"] + 1):
    ckpt = f"{cfg_G['name']}_ep{ep}"
    if (CHECKPOINT_DIR / ckpt).exists():
        print(f"\nInference on {ckpt}...")
        r = run_inference_from_checkpoint(ckpt, cfg_G["prompt_fn"], desc=f"G_ep{ep}")
        G_epoch_results.append(r)
    else:
        print(f"Checkpoint {ckpt} not found — skipping.")

final_rpath = CHECKPOINT_DIR / f"{cfg_G['name']}_final_results.json"
if final_rpath.exists():
    with open(final_rpath) as f:
        G_epoch_results.append(json.load(f))

print(f"\nModel G: {len(G_epoch_results)} checkpoints ready.")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 14: Combine with v2 results + Build Ensembles
# ══════════════════════════════════════════════════════════════════════════════

# Load v2 results
V2_DIR = PROJECT_ROOT / "checkpoints_v2"
v2_models = ["A_cap_meta_dora_r8a16_s42", "B_cap_meta_dora_r8a16_s123",
             "C_cap_sol_dora_r8a16_s456", "D_cap_meta_dora_r8a32_s789",
             "E_nocap_lora_r8a16_s42"]

all_results = []
for name in v2_models:
    rpath = V2_DIR / f"{name}_results.json"
    if rpath.exists():
        with open(rpath) as f:
            all_results.append(json.load(f))
        print(f"Loaded v2: {name}")

# Add F and G results (all epochs + finals)
for r in F_epoch_results:
    all_results.append(r)
    print(f"Added: {r['name']}")

for r in G_epoch_results:
    all_results.append(r)
    print(f"Added: {r['name']}")

print(f"\nTotal models for ensemble: {len(all_results)}")

# ── Build ensembles ──
test_nc = [len(row["choices"]) for _, row in test_df.iterrows()]

def score_avg_ensemble(results_list, nc_list):
    score_lists = [r["test_scores"] for r in results_list]
    preds = []
    for i in range(len(score_lists[0])):
        avg = [np.mean([sl[i][ci] for sl in score_lists if ci < len(sl[i])])
               for ci in range(nc_list[i])]
        preds.append(int(np.argmax(avg)))
    return preds

# Ensemble: all models
all_preds = score_avg_ensemble(all_results, test_nc)
sub_all = pd.DataFrame({"id": test_df["id"], "answer": all_preds})
sub_all.to_csv(SUBMISSION_DIR / "sub_ensemble_all_FG.csv", index=False)

# Ensemble: v2 (A-E) + F_final + G_final only
v2_plus_fg = [r for r in all_results if any(x in r["name"] for x in v2_models + ["_final"])]
fg_preds = score_avg_ensemble(v2_plus_fg, test_nc)
sub_fg = pd.DataFrame({"id": test_df["id"], "answer": fg_preds})
sub_fg.to_csv(SUBMISSION_DIR / "sub_ensemble_v2_FG_finals.csv", index=False)

# Ensemble: drop C (known bad)
no_c = [r for r in all_results if "cap_sol" not in r["name"] and "C_" not in r["name"]]
noc_preds = score_avg_ensemble(no_c, test_nc)
sub_noc = pd.DataFrame({"id": test_df["id"], "answer": noc_preds})
sub_noc.to_csv(SUBMISSION_DIR / "sub_ensemble_all_noC.csv", index=False)

print(f"\nGenerated ensembles:")
print(f"  sub_ensemble_all_FG.csv ({len(all_results)} models)")
print(f"  sub_ensemble_v2_FG_finals.csv ({len(v2_plus_fg)} models)")
print(f"  sub_ensemble_all_noC.csv ({len(no_c)} models)")
print(f"\nAll saved to {SUBMISSION_DIR}")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 15: Summary
# ══════════════════════════════════════════════════════════════════════════════
print(f"\n{'='*60}")
print(f"NOTEBOOK 1 COMPLETE")
print(f"{'='*60}")
print(f"Models trained: F ({cfg_F['epochs']} epochs) + G ({cfg_G['epochs']} epochs)")
print(f"Epoch checkpoints inferred: {len(F_epoch_results) + len(G_epoch_results)}")
print(f"Total ensemble models: {len(all_results)}")
print(f"\nSubmissions at: {SUBMISSION_DIR}")
for f in sorted(os.listdir(SUBMISSION_DIR)):
    if f.endswith('.csv'):
        print(f"  {f}")
print(f"\nUpload ensemble CSVs to Kaggle!")
```
