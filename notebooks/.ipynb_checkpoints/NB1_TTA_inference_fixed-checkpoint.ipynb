# NOTEBOOK 1: Full Permutation TTA on Existing 5 Models
=====================================================
NO TRAINING. Just re-run inference with every permutation of choice order.
Removes positional bias → free +1-2%.

Run on A100. ~2 hours for all 5 models × full permutations.

Expected output: submission_tta_ensemble.csv


```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 1: Setup
```

```python
# ══════════════════════════════════════════════════════════════════════════════
from google.colab import drive
drive.mount('/content/drive')

import os, json, gc, time, random
from pathlib import Path
from itertools import permutations
import numpy as np
import pandas as pd
from PIL import Image
import torch
from tqdm import tqdm

PROJECT_ROOT = Path("/content/drive/MyDrive/kaggle_final_competition")
DATA_DIR = PROJECT_ROOT / "data"
V2_DIR = PROJECT_ROOT / "checkpoints_v2"
CAPTION_DIR = PROJECT_ROOT / "captions"
SUBMISSION_DIR = PROJECT_ROOT / "submissions_v4"
RESULTS_DIR = PROJECT_ROOT / "results_v4"

for d in [SUBMISSION_DIR, RESULTS_DIR]:
    d.mkdir(parents=True, exist_ok=True)

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 2: Install + Imports
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# !pip install -q transformers==4.57.6 peft==0.18.1 bitsandbytes accelerate datasets pillow

from transformers import AutoProcessor, AutoModelForVision2Seq, BitsAndBytesConfig
from peft import PeftModel
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 3: Load Data + Captions
```

```python
# ══════════════════════════════════════════════════════════════════════════════
train_df = pd.read_csv(DATA_DIR / "train.csv")
val_df   = pd.read_csv(DATA_DIR / "val.csv")
test_df  = pd.read_csv(DATA_DIR / "test.csv")

for df in [train_df, val_df, test_df]:
    df["choices"] = df["choices"].apply(json.loads)

# Load captions
all_captions = {}
for split in ["train", "val", "test"]:
    cap_path = CAPTION_DIR / f"{split}_captions.json"
    if cap_path.exists():
        with open(cap_path) as f:
            all_captions.update(json.load(f))
print(f"Loaded {len(all_captions)} captions")

CHOICE_LETTERS = "ABCDEFGHIJ"
MODEL_ID = "HuggingFaceTB/SmolVLM-500M-Instruct"
MAX_CONTEXT_CHARS = 400

# Image resolution knob.
# SmolVLM expands every <image> placeholder into many visual tokens. Default
# (longest_edge = N*384 with N=4 = 1536) yields ~1100 image tokens per sample.
# For full-permutation TTA (up to 24 forward passes per sample), this is the
# main lever for both speed and peak memory. Lower to 768 if you hit OOM or
# want a faster sweep; the inference-time accuracy hit is small.
IMAGE_LONGEST_EDGE = 1536  # set 768 for faster/lighter TTA
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 4: Prompt Builders (same as v2 training)
```

```python
# ══════════════════════════════════════════════════════════════════════════════
def build_prompt_caption_meta(row, include_answer=False):
    """Caption + metadata prompt (used by models A, B, D)"""
    parts = []
    # Caption
    cap = all_captions.get(row["id"], "")
    if cap:
        parts.append(f"Image description: {cap[:300]}")
    # Metadata
    if pd.notna(row.get("subject", "")):
        parts.append(f"Subject: {row['subject']}")
    if pd.notna(row.get("topic", "")):
        parts.append(f"Topic: {row['topic']}")
    # Lecture + hint
    lecture = row.get("lecture", "")
    hint = row.get("hint", "")
    if pd.notna(lecture) and str(lecture).strip():
        parts.append(str(lecture).strip()[:MAX_CONTEXT_CHARS])
    if pd.notna(hint) and str(hint).strip():
        parts.append(str(hint).strip()[:MAX_CONTEXT_CHARS])

    context = "\n".join(parts)
    choices = row["choices"]
    choices_str = "\n".join(f"  {CHOICE_LETTERS[i]}. {c}" for i, c in enumerate(choices))

    prompt = "<image>\n"
    if context:
        prompt += f"Context:\n{context}\n\n"
    prompt += f"Question: {row['question']}\n"
    prompt += f"Choices:\n{choices_str}\n"
    prompt += "Answer:"
    if include_answer:
        answer_idx = int(row['answer'])
        prompt += f" {CHOICE_LETTERS[answer_idx]}. {choices[answer_idx]}"
    return prompt


def build_prompt_caption_solution(row, include_answer=False):
    """Caption + solution prompt (used by model C)"""
    parts = []
    cap = all_captions.get(row["id"], "")
    if cap:
        parts.append(f"Image description: {cap[:300]}")
    lecture = row.get("lecture", "")
    hint = row.get("hint", "")
    if pd.notna(lecture) and str(lecture).strip():
        parts.append(str(lecture).strip()[:MAX_CONTEXT_CHARS])
    if pd.notna(hint) and str(hint).strip():
        parts.append(str(hint).strip()[:MAX_CONTEXT_CHARS])

    context = "\n".join(parts)
    choices = row["choices"]
    choices_str = "\n".join(f"  {CHOICE_LETTERS[i]}. {c}" for i, c in enumerate(choices))

    prompt = "<image>\n"
    if context:
        prompt += f"Context:\n{context}\n\n"
    prompt += f"Question: {row['question']}\n"
    prompt += f"Choices:\n{choices_str}\n"
    prompt += "Answer:"
    if include_answer:
        answer_idx = int(row['answer'])
        sol = row.get("solution", "")
        if pd.notna(sol) and str(sol).strip():
            prompt += f" {CHOICE_LETTERS[answer_idx]}. {choices[answer_idx]}\nExplanation: {str(sol).strip()[:300]}"
        else:
            prompt += f" {CHOICE_LETTERS[answer_idx]}. {choices[answer_idx]}"
    return prompt


def build_prompt_nocap(row, include_answer=False):
    """No caption prompt (used by model E)"""
    parts = []
    lecture = row.get("lecture", "")
    hint = row.get("hint", "")
    if pd.notna(lecture) and str(lecture).strip():
        parts.append(str(lecture).strip()[:MAX_CONTEXT_CHARS])
    if pd.notna(hint) and str(hint).strip():
        parts.append(str(hint).strip()[:MAX_CONTEXT_CHARS])

    context = "\n".join(parts)
    choices = row["choices"]
    choices_str = "\n".join(f"  {CHOICE_LETTERS[i]}. {c}" for i, c in enumerate(choices))

    prompt = "<image>\n"
    if context:
        prompt += f"Context:\n{context}\n\n"
    prompt += f"Question: {row['question']}\n"
    prompt += f"Choices:\n{choices_str}\n"
    prompt += "Answer:"
    if include_answer:
        answer_idx = int(row['answer'])
        prompt += f" {CHOICE_LETTERS[answer_idx]}. {choices[answer_idx]}"
    return prompt


# Model configs — maps model name to its prompt function
MODEL_CONFIGS = {
    "A_cap_meta_dora_r8a16_s42":  build_prompt_caption_meta,
    "B_cap_meta_dora_r8a16_s123": build_prompt_caption_meta,
    "C_cap_sol_dora_r8a16_s456":  build_prompt_caption_solution,
    "D_cap_meta_dora_r8a32_s789": build_prompt_caption_meta,
    "E_nocap_lora_r8a16_s42":     build_prompt_nocap,
}
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 5: Multi-form token scoring (from v2)
```

```python
# ══════════════════════════════════════════════════════════════════════════════
def get_candidate_token_ids(tokenizer):
    """Get token IDs for each answer letter in multiple forms."""
    candidates = {}
    for letter in CHOICE_LETTERS[:10]:
        tids = set()
        for form in [letter, f" {letter}", f"{letter}.", f" {letter}."]:
            ids = tokenizer.encode(form, add_special_tokens=False)
            tids.update(ids)
        candidates[letter] = list(tids)
    return candidates
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 6: Core TTA inference function
```

```python
# ══════════════════════════════════════════════════════════════════════════════
def score_single_sample(model, processor, row, prompt_fn, candidate_ids, img_size=None):
    """Score a single sample. Returns list of log-probs per choice."""
    # Load image
    img = Image.open(DATA_DIR / row["image_path"]).convert("RGB")
    if img_size:
        img = img.resize((img_size, img_size), Image.BICUBIC)

    prompt = prompt_fn(row, include_answer=False)
    inputs = processor(text=[prompt], images=[img], return_tensors="pt")
    inputs = {k: v.to(model.device) if torch.is_tensor(v) else v for k, v in inputs.items()}

    with torch.inference_mode():
        out = model(**inputs)
        logits = out.logits[0]  # (seq_len, vocab)

    # Find last non-pad token
    input_ids = inputs["input_ids"][0]
    pad_id = processor.tokenizer.pad_token_id
    if pad_id is not None:
        non_pad = (input_ids != pad_id).nonzero(as_tuple=True)[0]
        last_pos = non_pad[-1].item()
    else:
        last_pos = len(input_ids) - 1

    log_probs = torch.log_softmax(logits[last_pos].float(), dim=-1)

    scores = []
    for ci in range(len(row["choices"])):
        letter = CHOICE_LETTERS[ci]
        tids = candidate_ids.get(letter, [])
        if tids:
            scores.append(max(log_probs[tid].item() for tid in tids))
        else:
            scores.append(-999.0)

    return scores


def run_full_permutation_tta(model, processor, df, prompt_fn, candidate_ids,
                              desc="TTA", max_perms=24):
    """
    Full permutation TTA:
    - For each sample, generate ALL permutations of choice order
    - Score each permutation
    - Map scores back to original order and average
    
    For 2-choice: 2 perms, 3-choice: 6 perms, 4-choice: 24 perms
    """
    all_scores = []
    all_preds = []

    for idx in tqdm(range(len(df)), desc=desc):
        row = df.iloc[idx]
        n_choices = len(row["choices"])
        original_choices = row["choices"]

        # Generate all permutations (capped at max_perms)
        all_perms = list(permutations(range(n_choices)))
        if len(all_perms) > max_perms:
            # For 5+ choices, sample random perms
            rng = random.Random(42 + idx)
            all_perms = [all_perms[0]] + rng.sample(all_perms[1:], max_perms - 1)

        # Accumulate scores in original order
        accum = np.zeros(n_choices)
        n_valid = 0

        for perm in all_perms:
            # Create shuffled row
            shuffled_row = row.copy()
            shuffled_row["choices"] = [original_choices[j] for j in perm]

            # Score with shuffled order
            shuffled_scores = score_single_sample(
                model, processor, shuffled_row, prompt_fn, candidate_ids
            )

            # Map back to original positions
            original_scores = np.full(n_choices, -999.0)
            for new_pos, old_pos in enumerate(perm):
                original_scores[old_pos] = shuffled_scores[new_pos]

            accum += original_scores
            n_valid += 1

        avg_scores = (accum / n_valid).tolist()
        pred = int(np.argmax(avg_scores))

        all_scores.append(avg_scores)
        all_preds.append(pred)

    return all_preds, all_scores
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 7: Load each model and run TTA
```

```python
# ══════════════════════════════════════════════════════════════════════════════
def _apply_image_size(processor, longest_edge):
    """Configure the processor's image size in a way that survives across
    transformers versions (the attribute path moved a few times)."""
    new_size = {"longest_edge": longest_edge}
    try:
        processor.image_processor.size = new_size
    except Exception:
        pass
    if hasattr(processor, "size"):
        try:
            processor.size = new_size
        except Exception:
            pass
    return processor


def find_adapter_dir(base_dir, model_name):
    """Resolve the actual adapter directory.

    NB2 saves checkpoints as ``{exp_name}_ep{N}``. ``V2_DIR / model_name`` itself
    rarely contains ``adapter_config.json``. Strategy:
      1. If ``adapter_config.json`` is directly under ``V2_DIR / model_name``, use it.
      2. Otherwise look for ``{model_name}_ep*`` siblings inside ``V2_DIR`` (NB2 layout)
         and pick the one with the highest epoch.
      3. Otherwise look for any subdirectory under ``V2_DIR / model_name`` that has
         ``adapter_config.json`` and pick the one with the highest epoch number.
    """
    direct = base_dir / model_name
    if (direct / "adapter_config.json").exists():
        return direct

    sibling_candidates = sorted(base_dir.glob(f"{model_name}_ep*"))
    sibling_candidates = [p for p in sibling_candidates if (p / "adapter_config.json").exists()]
    if sibling_candidates:
        def ep_num(p):
            try:
                return int(p.name.rsplit("_ep", 1)[-1])
            except ValueError:
                return -1
        return max(sibling_candidates, key=ep_num)

    if direct.exists():
        nested = [p for p in direct.iterdir()
                  if p.is_dir() and (p / "adapter_config.json").exists()]
        if nested:
            def ep_num(p):
                name = p.name
                try:
                    return int(name.rsplit("ep", 1)[-1])
                except ValueError:
                    return -1
            return max(nested, key=ep_num)

    raise FileNotFoundError(
        f"Could not find an adapter for '{model_name}' under {base_dir}. "
        f"Looked for {direct}/adapter_config.json, {base_dir}/{model_name}_ep*, "
        f"and nested subfolders."
    )


tta_results = []

for model_name, prompt_fn in MODEL_CONFIGS.items():
    tta_path = RESULTS_DIR / f"tta_{model_name}.json"

    # CACHE: skip if already done
    if tta_path.exists():
        print(f"CACHED: {model_name}")
        with open(tta_path) as f:
            tta_results.append(json.load(f))
        continue

    print(f"\n{'='*60}")
    print(f"Loading {model_name}...")
    t0 = time.time()

    bnb_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_compute_dtype=torch.float16,
        bnb_4bit_use_double_quant=True,
    )

    processor = AutoProcessor.from_pretrained(MODEL_ID)
    if processor.tokenizer.pad_token is None:
        processor.tokenizer.pad_token = processor.tokenizer.eos_token
    _apply_image_size(processor, IMAGE_LONGEST_EDGE)

    base_model = AutoModelForVision2Seq.from_pretrained(
        MODEL_ID,
        quantization_config=bnb_config,
        device_map="auto",
        low_cpu_mem_usage=True,
    )
    # Inference-only — keep KV cache enabled (the default).

    # Load LoRA adapter — resolve the real checkpoint path
    adapter_path = find_adapter_dir(V2_DIR, model_name)
    print(f"  Using adapter: {adapter_path}")
    model = PeftModel.from_pretrained(base_model, str(adapter_path))
    model.eval()

    candidate_ids = get_candidate_token_ids(processor.tokenizer)

    # ── Val TTA (to measure improvement) ──
    print(f"Running val TTA...")
    val_preds_tta, val_scores_tta = run_full_permutation_tta(
        model, processor, val_df, prompt_fn, candidate_ids,
        desc=f"Val TTA {model_name[:20]}"
    )
    val_correct = sum(p == int(val_df.iloc[i]["answer"])
                      for i, p in enumerate(val_preds_tta))
    val_acc_tta = val_correct / len(val_df)
    print(f"Val accuracy (full perm TTA): {val_acc_tta:.4f}")

    # ── Test TTA ──
    print(f"Running test TTA...")
    test_preds_tta, test_scores_tta = run_full_permutation_tta(
        model, processor, test_df, prompt_fn, candidate_ids,
        desc=f"Test TTA {model_name[:20]}"
    )

    result = {
        "name": model_name,
        "val_acc_tta": val_acc_tta,
        "val_preds_tta": val_preds_tta,
        "val_scores_tta": val_scores_tta,
        "test_preds_tta": test_preds_tta,
        "test_scores_tta": test_scores_tta,
        "time_min": (time.time() - t0) / 60,
    }
    tta_results.append(result)

    with open(tta_path, "w") as f:
        json.dump(result, f)
    print(f"Saved: {tta_path.name} ({result['time_min']:.1f} min)")

    # Save individual submission
    sub = pd.DataFrame({"id": test_df["id"], "answer": test_preds_tta})
    sub.to_csv(SUBMISSION_DIR / f"sub_tta_{model_name}.csv", index=False)

    del model, base_model, processor
    gc.collect()
    torch.cuda.empty_cache()
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 8: TTA Ensemble
```

```python
# ══════════════════════════════════════════════════════════════════════════════
print(f"\n{'='*60}")
print("TTA ENSEMBLE RESULTS")
print(f"{'='*60}")

for r in tta_results:
    print(f"  {r['name'][:40]}: val_tta={r['val_acc_tta']:.4f}")

# Score average ensemble (TTA scores)
test_nc = [len(row["choices"]) for _, row in test_df.iterrows()]

def score_avg_ensemble(results_list, num_choices):
    preds = []
    for i in range(len(num_choices)):
        avg = []
        for ci in range(num_choices[i]):
            vals = [r["test_scores_tta"][i][ci] for r in results_list
                    if ci < len(r["test_scores_tta"][i])]
            avg.append(np.mean(vals))
        preds.append(int(np.argmax(avg)))
    return preds

# All 5 TTA
tta_all_preds = score_avg_ensemble(tta_results, test_nc)
sub_all = pd.DataFrame({"id": test_df["id"], "answer": tta_all_preds})
sub_all.to_csv(SUBMISSION_DIR / "sub_tta_ensemble_all5.csv", index=False)

# Without C (known weak)
tta_no_c = [r for r in tta_results if "C_" not in r["name"] and "cap_sol" not in r["name"]]
tta_noc_preds = score_avg_ensemble(tta_no_c, test_nc)
sub_noc = pd.DataFrame({"id": test_df["id"], "answer": tta_noc_preds})
sub_noc.to_csv(SUBMISSION_DIR / "sub_tta_ensemble_no_C.csv", index=False)

# Val ensemble accuracy
val_nc = [len(row["choices"]) for _, row in val_df.iterrows()]
def val_ensemble_acc(results_list):
    preds = []
    for i in range(len(val_nc)):
        avg = []
        for ci in range(val_nc[i]):
            vals = [r["val_scores_tta"][i][ci] for r in results_list
                    if ci < len(r["val_scores_tta"][i])]
            avg.append(np.mean(vals))
        preds.append(int(np.argmax(avg)))
    correct = sum(p == int(val_df.iloc[i]["answer"]) for i, p in enumerate(preds))
    return correct / len(val_df)

print(f"\nTTA Ensemble (all 5) val acc: {val_ensemble_acc(tta_results):.4f}")
print(f"TTA Ensemble (no C)  val acc: {val_ensemble_acc(tta_no_c):.4f}")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 9: Also combine original (non-TTA) + TTA scores
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Load original v2 results
v2_results = []
for name in MODEL_CONFIGS.keys():
    v2_path = V2_DIR / f"{name}_results.json"
    if v2_path.exists():
        with open(v2_path) as f:
            v2_results.append(json.load(f))

# Combined: average original scores + TTA scores (double the votes)
combined_preds = []
for i in range(len(test_nc)):
    avg = []
    for ci in range(test_nc[i]):
        vals = []
        # Original scores
        for r in v2_results:
            if ci < len(r["test_scores"][i]):
                vals.append(r["test_scores"][i][ci])
        # TTA scores
        for r in tta_results:
            if ci < len(r["test_scores_tta"][i]):
                vals.append(r["test_scores_tta"][i][ci])
        avg.append(np.mean(vals) if vals else -999.0)
    combined_preds.append(int(np.argmax(avg)))

sub_combined = pd.DataFrame({"id": test_df["id"], "answer": combined_preds})
sub_combined.to_csv(SUBMISSION_DIR / "sub_tta_plus_original_combined.csv", index=False)

print(f"\nGenerated submissions:")
for f in sorted(os.listdir(SUBMISSION_DIR)):
    if f.endswith(".csv"):
        print(f"  {f}")

print(f"\n{'='*60}")
print("NOTEBOOK 1 COMPLETE — upload sub_tta_ensemble_all5.csv first")
print(f"{'='*60}")
```
