# NOTEBOOK 2: Pseudo-Label + Retrain New Models
==============================================
1. Use existing 5-model ensemble to pseudo-label high-confidence test samples
2. Train 2 NEW models on train + val + pseudo-labels (more data!)
3. One model uses curriculum learning (easy → hard)

Run on A100 #2 in parallel with Notebook 1 (TTA).
~3-4 hours total.


```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 1: Setup
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# from google.colab import drive
# drive.mount('/content/drive')

import os, json, gc, time, random, math
from pathlib import Path
import numpy as np
import pandas as pd
from PIL import Image
import torch
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm

PROJECT_ROOT = Path("/scratch/au2414/kaggle_final_competition/")
DATA_DIR = PROJECT_ROOT / "data"
V2_DIR = PROJECT_ROOT / "checkpoints_v2"
CAPTION_DIR = PROJECT_ROOT / "captions"
CHECKPOINT_DIR = PROJECT_ROOT / "checkpoints_v4"
SUBMISSION_DIR = PROJECT_ROOT / "submissions_v4"
RESULTS_DIR = PROJECT_ROOT / "results_v4"

for d in [CHECKPOINT_DIR, SUBMISSION_DIR, RESULTS_DIR]:
    d.mkdir(parents=True, exist_ok=True)

device = torch.device("cuda")
print(f"GPU: {torch.cuda.get_device_name(0)}")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 2: Install
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# !pip install -q transformers==4.57.6 peft==0.18.1 bitsandbytes accelerate datasets pillow

from transformers import AutoProcessor, AutoModelForVision2Seq, BitsAndBytesConfig
from peft import LoraConfig, get_peft_model, PeftModel, TaskType
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 3: Load Data + Captions
```

```python
# ══════════════════════════════════════════════════════════════════════════════
train_df = pd.read_csv(DATA_DIR / "train.csv")
val_df   = pd.read_csv(DATA_DIR / "val.csv")
test_df  = pd.read_csv(DATA_DIR / "test.csv")

for df in [train_df, val_df, test_df]:
    df["choices"] = df["choices"].apply(json.loads)

all_captions = {}
for split in ["train", "val", "test"]:
    cap_path = CAPTION_DIR / f"{split}_captions.json"
    if cap_path.exists():
        with open(cap_path) as f:
            all_captions.update(json.load(f))
print(f"Loaded {len(all_captions)} captions")

CHOICE_LETTERS = "ABCDEFGHIJ"
MODEL_ID = "HuggingFaceTB/SmolVLM-500M-Instruct"
MAX_CONTEXT_CHARS = 400
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 4: Generate Pseudo-Labels from v2 ensemble
```

```python
# ══════════════════════════════════════════════════════════════════════════════
print("Loading v2 results to generate pseudo-labels...")

v2_model_names = [
    "A_cap_meta_dora_r8a16_s42",
    "B_cap_meta_dora_r8a16_s123",
    "C_cap_sol_dora_r8a16_s456",
    "D_cap_meta_dora_r8a32_s789",
    "E_nocap_lora_r8a16_s42",
]

v2_results = []
for name in v2_model_names:
    rpath = V2_DIR / f"{name}_results.json"
    if rpath.exists():
        with open(rpath) as f:
            v2_results.append(json.load(f))

print(f"Loaded {len(v2_results)} v2 model results")

# Score-average ensemble on test
test_nc = [len(row["choices"]) for _, row in test_df.iterrows()]
pseudo_labels = []
pseudo_confidences = []
pseudo_agreements = []

for i in range(len(test_df)):
    nc = test_nc[i]
    avg_scores = []
    for ci in range(nc):
        vals = [r["test_scores"][i][ci] for r in v2_results if ci < len(r["test_scores"][i])]
        avg_scores.append(np.mean(vals))

    # Softmax to get probabilities
    scores_arr = np.array(avg_scores)
    exp_scores = np.exp(scores_arr - np.max(scores_arr))
    probs = exp_scores / exp_scores.sum()

    pred = int(np.argmax(probs))
    confidence = probs[pred]

    # Count agreement (how many models agree)
    individual_preds = []
    for r in v2_results:
        s = r["test_scores"][i][:nc]
        individual_preds.append(int(np.argmax(s)))
    agreement = sum(1 for p in individual_preds if p == pred) / len(individual_preds)

    pseudo_labels.append(pred)
    pseudo_confidences.append(confidence)
    pseudo_agreements.append(agreement)

# Filter: only keep high-confidence pseudo-labels
# Require: confidence > 0.8 AND all 5 models agree
CONF_THRESHOLD = 0.8
AGREEMENT_THRESHOLD = 1.0  # unanimous

pseudo_mask = [(c > CONF_THRESHOLD and a >= AGREEMENT_THRESHOLD)
               for c, a in zip(pseudo_confidences, pseudo_agreements)]
n_pseudo = sum(pseudo_mask)
print(f"\nPseudo-label stats:")
print(f"  Total test samples: {len(test_df)}")
print(f"  High confidence (>{CONF_THRESHOLD}): {sum(c > CONF_THRESHOLD for c in pseudo_confidences)}")
print(f"  Unanimous agreement: {sum(a >= AGREEMENT_THRESHOLD for a in pseudo_agreements)}")
print(f"  Both (final pseudo-labels): {n_pseudo} ({n_pseudo/len(test_df)*100:.1f}%)")

# Create pseudo-label dataframe
pseudo_rows = []
for i in range(len(test_df)):
    if pseudo_mask[i]:
        row = test_df.iloc[i].copy()
        row["answer"] = pseudo_labels[i]
        pseudo_rows.append(row)

pseudo_df = pd.DataFrame(pseudo_rows)
print(f"Pseudo-label DF: {len(pseudo_df)} rows")

# Combined training data: train + val + pseudo
combined_df = pd.concat([train_df, val_df, pseudo_df], ignore_index=True)
print(f"Combined training data: {len(train_df)} train + {len(val_df)} val + {len(pseudo_df)} pseudo = {len(combined_df)} total")
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 5: Prompt Builder + Dataset
```

```python
# ══════════════════════════════════════════════════════════════════════════════
def build_prompt_caption_meta(row, include_answer=False):
    parts = []
    cap = all_captions.get(row["id"], "")
    if cap:
        parts.append(f"Image description: {cap[:300]}")
    if pd.notna(row.get("subject", "")):
        parts.append(f"Subject: {row['subject']}")
    if pd.notna(row.get("topic", "")):
        parts.append(f"Topic: {row['topic']}")
    lecture = row.get("lecture", "")
    hint = row.get("hint", "")
    if pd.notna(lecture) and str(lecture).strip():
        parts.append(str(lecture).strip()[:MAX_CONTEXT_CHARS])
    if pd.notna(hint) and str(hint).strip():
        parts.append(str(hint).strip()[:MAX_CONTEXT_CHARS])

    context = "\n".join(parts)
    choices = row["choices"]
    choices_str = "\n".join(f"  {CHOICE_LETTERS[i]}. {c}" for i, c in enumerate(choices))

    prompt = "<image>\n"
    if context:
        prompt += f"Context:\n{context}\n\n"
    prompt += f"Question: {row['question']}\n"
    prompt += f"Choices:\n{choices_str}\n"
    prompt += "Answer:"
    if include_answer:
        answer_idx = int(row['answer'])
        prompt += f" {CHOICE_LETTERS[answer_idx]}. {choices[answer_idx]}"
    return prompt


def get_candidate_token_ids(tokenizer):
    candidates = {}
    for letter in CHOICE_LETTERS[:10]:
        tids = set()
        for form in [letter, f" {letter}", f"{letter}.", f" {letter}."]:
            ids = tokenizer.encode(form, add_special_tokens=False)
            tids.update(ids)
        candidates[letter] = list(tids)
    return candidates
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 6: Training Infrastructure
```

```python
# ══════════════════════════════════════════════════════════════════════════════
def load_fresh_model(seed, lora_r=8, lora_alpha=16, use_dora=True):
    """Load a fresh base model + apply new LoRA."""
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    bnb_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_compute_dtype=torch.float16,
        bnb_4bit_use_double_quant=True,
    )

    processor = AutoProcessor.from_pretrained(MODEL_ID)
    if processor.tokenizer.pad_token is None:
        processor.tokenizer.pad_token = processor.tokenizer.eos_token

    base_model = AutoModelForVision2Seq.from_pretrained(
        MODEL_ID,
        quantization_config=bnb_config,
        device_map="auto",
        low_cpu_mem_usage=True,
    )

    lora_config = LoraConfig(
        r=lora_r,
        lora_alpha=lora_alpha,
        target_modules=[
            "q_proj", "k_proj", "v_proj", "o_proj",
            "gate_proj", "up_proj", "down_proj"
        ],
        lora_dropout=0.05,
        bias="none",
        task_type=TaskType.CAUSAL_LM,
        use_dora=use_dora,
    )

    model = get_peft_model(base_model, lora_config)
    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"Trainable params: {trainable:,} ({trainable/1e6:.2f}M)")

    return model, processor


def train_collate_fn(batch, processor):
    """Left-padded collation with answer-only loss masking."""
    images = [b["image"] for b in batch]
    texts = [b["text"] for b in batch]

    encodings = processor(text=texts, images=images, return_tensors="pt",
                          padding=True, truncation=True, max_length=512)

    input_ids = encodings["input_ids"]
    labels = input_ids.clone()

    # Mask everything except the answer part
    pad_id = processor.tokenizer.pad_token_id
    for i in range(len(batch)):
        prompt = build_prompt_caption_meta(batch[i]["row"], include_answer=False)
        prompt_ids = processor.tokenizer.encode(prompt, add_special_tokens=False)
        prompt_len = len(prompt_ids)

        # Find where actual tokens start (skip padding)
        non_pad = (input_ids[i] != pad_id).nonzero(as_tuple=True)[0]
        if len(non_pad) > 0:
            start = non_pad[0].item()
            # Mask pad tokens and prompt tokens
            labels[i, :start + prompt_len] = -100
        labels[i, input_ids[i] == pad_id] = -100

    encodings["labels"] = labels
    return encodings


def train_model(model, processor, df, prompt_fn, exp_name, seed,
                epochs=2, lr=3e-5, batch_size=1, grad_accum=8,
                use_cosine=True, curriculum=False):
    """Train with optional curriculum learning."""
    random.seed(seed)

    # Prepare training data
    train_items = []
    for _, row in df.iterrows():
        img = Image.open(DATA_DIR / row["image_path"]).convert("RGB")
        text = prompt_fn(row, include_answer=True)
        train_items.append({"image": img, "text": text, "row": row})

    if curriculum:
        # Sort by text length (proxy for difficulty — shorter = easier)
        train_items.sort(key=lambda x: len(x["text"]))
        print(f"Curriculum: sorted {len(train_items)} items by difficulty")

    model.train()
    optimizer = torch.optim.AdamW(
        [p for p in model.parameters() if p.requires_grad],
        lr=lr, weight_decay=0.01
    )

    total_steps = (len(train_items) * epochs) // (batch_size * grad_accum)
    warmup_steps = int(0.05 * total_steps)

    def get_lr(step):
        if step < warmup_steps:
            return step / max(warmup_steps, 1)
        if use_cosine:
            progress = (step - warmup_steps) / max(total_steps - warmup_steps, 1)
            return 0.5 * (1 + math.cos(math.pi * progress))
        return 1.0

    scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, get_lr)

    global_step = 0
    for epoch in range(epochs):
        if not curriculum:
            random.shuffle(train_items)

        epoch_loss = 0
        n_batches = 0
        pbar = tqdm(range(0, len(train_items), batch_size),
                     desc=f"Epoch {epoch+1}/{epochs}")

        for bi in pbar:
            batch = train_items[bi:bi+batch_size]
            encodings = train_collate_fn(batch, processor)
            encodings = {k: v.to(model.device) if torch.is_tensor(v) else v
                        for k, v in encodings.items()}

            out = model(**encodings)
            loss = out.loss / grad_accum
            loss.backward()

            epoch_loss += out.loss.item()
            n_batches += 1

            if (n_batches % grad_accum == 0) or (bi + batch_size >= len(train_items)):
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()
                scheduler.step()
                optimizer.zero_grad()
                global_step += 1

            pbar.set_postfix(loss=f"{epoch_loss/n_batches:.4f}",
                            lr=f"{scheduler.get_last_lr()[0]:.2e}")

        avg_loss = epoch_loss / n_batches
        print(f"Epoch {epoch+1}: avg_loss={avg_loss:.4f}")

        # Save epoch checkpoint
        ckpt_path = CHECKPOINT_DIR / f"{exp_name}_ep{epoch+1}"
        model.save_pretrained(str(ckpt_path))
        print(f"Checkpoint: {ckpt_path.name}")

    return model


def run_inference(model, processor, df, prompt_fn, candidate_ids, desc="Eval"):
    """Standard log-likelihood inference (no TTA)."""
    model.eval()
    preds = []
    scores = []

    for idx in tqdm(range(len(df)), desc=desc):
        row = df.iloc[idx]
        img = Image.open(DATA_DIR / row["image_path"]).convert("RGB")
        prompt = prompt_fn(row, include_answer=False)

        inputs = processor(text=[prompt], images=[img], return_tensors="pt")
        inputs = {k: v.to(model.device) if torch.is_tensor(v) else v
                  for k, v in inputs.items()}

        with torch.inference_mode():
            out = model(**inputs)
            logits = out.logits[0]

        input_ids = inputs["input_ids"][0]
        pad_id = processor.tokenizer.pad_token_id
        if pad_id is not None:
            non_pad = (input_ids != pad_id).nonzero(as_tuple=True)[0]
            last_pos = non_pad[-1].item()
        else:
            last_pos = len(input_ids) - 1

        log_probs = torch.log_softmax(logits[last_pos].float(), dim=-1)

        sample_scores = []
        for ci in range(len(row["choices"])):
            letter = CHOICE_LETTERS[ci]
            tids = candidate_ids.get(letter, [])
            if tids:
                sample_scores.append(max(log_probs[tid].item() for tid in tids))
            else:
                sample_scores.append(-999.0)

        scores.append(sample_scores)
        preds.append(int(np.argmax(sample_scores)))

    return preds, scores
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 7: Train Model P1 (Pseudo-label, standard)
```

```python
# ══════════════════════════════════════════════════════════════════════════════
P1_NAME = "P1_pseudo_dora_r8a16_s42"
p1_results_path = RESULTS_DIR / f"{P1_NAME}_results.json"

if p1_results_path.exists():
    print(f"CACHED: {P1_NAME}")
    with open(p1_results_path) as f:
        p1_result = json.load(f)
else:
    print(f"\n{'='*60}")
    print(f"Training {P1_NAME}")
    print(f"Data: {len(combined_df)} samples (train+val+pseudo)")
    print(f"{'='*60}")

    model_p1, proc_p1 = load_fresh_model(seed=42, lora_r=8, lora_alpha=16, use_dora=True)

    model_p1 = train_model(
        model_p1, proc_p1, combined_df, build_prompt_caption_meta,
        exp_name=P1_NAME, seed=42, epochs=2, lr=3e-5,
        batch_size=1, grad_accum=8, use_cosine=True, curriculum=False,
    )

    candidate_ids = get_candidate_token_ids(proc_p1.tokenizer)

    print("Test inference P1...")
    p1_preds, p1_scores = run_inference(
        model_p1, proc_p1, test_df, build_prompt_caption_meta, candidate_ids,
        desc="Test P1"
    )

    p1_result = {
        "name": P1_NAME,
        "test_preds": p1_preds,
        "test_scores": p1_scores,
    }
    with open(p1_results_path, "w") as f:
        json.dump(p1_result, f)

    sub = pd.DataFrame({"id": test_df["id"], "answer": p1_preds})
    sub.to_csv(SUBMISSION_DIR / f"sub_{P1_NAME}.csv", index=False)
    print(f"Saved sub_{P1_NAME}.csv")

    del model_p1, proc_p1
    gc.collect()
    torch.cuda.empty_cache()
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 8: Train Model P2 (Pseudo-label + curriculum)
```

```python
# ══════════════════════════════════════════════════════════════════════════════
P2_NAME = "P2_pseudo_curriculum_dora_r8a16_s123"
p2_results_path = RESULTS_DIR / f"{P2_NAME}_results.json"

if p2_results_path.exists():
    print(f"CACHED: {P2_NAME}")
    with open(p2_results_path) as f:
        p2_result = json.load(f)
else:
    print(f"\n{'='*60}")
    print(f"Training {P2_NAME}")
    print(f"Data: {len(combined_df)} samples + curriculum ordering")
    print(f"{'='*60}")

    model_p2, proc_p2 = load_fresh_model(seed=123, lora_r=8, lora_alpha=16, use_dora=True)

    model_p2 = train_model(
        model_p2, proc_p2, combined_df, build_prompt_caption_meta,
        exp_name=P2_NAME, seed=123, epochs=3, lr=2e-5,
        batch_size=1, grad_accum=8, use_cosine=True, curriculum=True,
    )

    candidate_ids = get_candidate_token_ids(proc_p2.tokenizer)

    print("Test inference P2...")
    p2_preds, p2_scores = run_inference(
        model_p2, proc_p2, test_df, build_prompt_caption_meta, candidate_ids,
        desc="Test P2"
    )

    p2_result = {
        "name": P2_NAME,
        "test_preds": p2_preds,
        "test_scores": p2_scores,
    }
    with open(p2_results_path, "w") as f:
        json.dump(p2_result, f)

    sub = pd.DataFrame({"id": test_df["id"], "answer": p2_preds})
    sub.to_csv(SUBMISSION_DIR / f"sub_{P2_NAME}.csv", index=False)
    print(f"Saved sub_{P2_NAME}.csv")

    del model_p2, proc_p2
    gc.collect()
    torch.cuda.empty_cache()
```

```python
# ══════════════════════════════════════════════════════════════════════════════
# Cell 9: Train Model P3 (Pseudo-label + different rank + LR)
```

```python
# ══════════════════════════════════════════════════════════════════════════════
P3_NAME = "P3_pseudo_r16a32_dora_s456"
p3_results_path = RESULTS_DIR / f"{P3_NAME}_results.json"

if p3_results_path.exists():
    print(f"CACHED: {P3_NAME}")
    with open(p3_results_path) as f:
        p3_result = json.load(f)
else:
    print(f"\n{'='*60}")
    print(f"Training {P3_NAME}")
    print(f"Data: {len(combined_df)} samples, r=16, alpha=32")
    print(f"{'='*60}")

    model_p3, proc_p3 = load_fresh_model(seed=456, lora_r=16, lora_alpha=32, use_dora=True)

    model_p3 = train_model(
        model_p3, proc_p3, combined_df, build_prompt_caption_meta,
        exp_name=P3_NAME, seed=456, epochs=2, lr=3e-5,
        batch_size=1, grad_accum=8, use_cosine=True, curriculum=False,
    )

    candidate_ids = get_candidate_token_ids(proc_p3.tokenizer)

    print("Test inference P3...")
    p3_preds, p3_scores = run_inference(
        model_p3, proc_p3, test_df, build_prompt_caption_meta, candidate_ids,
        desc="Test P3"
    )

    p3_result = {
        "name": P3_NAME,
        "test_preds": p3_preds,
        "test_scores": p3_scores,
    }
    with open(p3_results_path, "w") as f:
        json.dump(p3_result, f)

    sub = pd.DataFrame({"id": test_df["id"], "answer": p3_preds})
    sub.to_csv(SUBMISSION_DIR / f"sub_{P3_NAME}.csv", index=False)
    print(f"Saved sub_{P3_NAME}.csv")

    del model_p3, proc_p3
    gc.collect()
    torch.cuda.empty_cache()

print(f"\n{'='*60}")
print(f"NOTEBOOK 2 COMPLETE — all pseudo-label models trained")
print(f"{'='*60}")
```
